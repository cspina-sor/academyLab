# Lab Terraform - Day 2
## Infrastructure as Code con Terraform su Azure

In questo laboratorio metteremo in pratica i concetti della Giornata 2: workflow `init/plan/apply/destroy`, gestione dello stato, utilizzo di variabili e moduli per distribuire una Web App dichiarativamente.

**Risultati attesi:**
- Preparare un repository Terraform leggibile.
- Riutilizzare il Resource Group dedicato della Giornata 1 senza ricrearlo (uso del blocco `data`).
- Deployare App Service Plan e Web App con variabili e naming coerenti.
- Leggere `plan`/`apply` per capire impatto e output.

### 1. Setup del progetto
Il codice è organizzato in una struttura modulare all'interno della cartella `Day2_Lab_Terraform`.

```powershell
Get-ChildItem -Recurse Day2_Lab_Terraform | Select-Object FullName
```

### 2. Autenticazione Azure
Prima di utilizzare Terraform, assicurati di essere autenticato sul tuo account Azure.

```powershell
az login
# Se hai più sottoscrizioni, seleziona quella corretta scommentando la riga seguente e inserendo l'ID:
# az account set --subscription "<YOUR_SUBSCRIPTION_ID>"
az account show
```

### 3. Inizializzazione (Init)
Il primo passo in qualsiasi progetto Terraform è l'inizializzazione. Questo scarica i provider (es. `azurerm`) e prepara il backend per lo state.

```powershell
Set-Location Day2_Lab_Terraform
terraform init
```

### 4. Validazione e Formattazione
Assicuriamoci che il codice sia scritto correttamente e ben formattato.

```powershell
Set-Location Day2_Lab_Terraform
terraform fmt -recursive
terraform validate
```

### 5. Preparazione delle Variabili
Crea un file `terraform.tfvars` a partire dall'esempio per passare i tuoi valori personalizzati (nome studente, nome del Resource Group della Giornata 1).

```powershell
Copy-Item "Day2_Lab_Terraform/terraform.tfvars.example" -Destination "Day2_Lab_Terraform/terraform.tfvars"
Get-Content Day2_Lab_Terraform/terraform.tfvars
```

### 6. Pianificazione (Plan)
Verifichiamo cosa farà Terraform. Notare che il Resource Group viene letto (tramite `data source`), mentre App Service Plan e Web App vengono creati.

```powershell
Set-Location Day2_Lab_Terraform
terraform plan -out=tfplan
```

### 7. Applicazione (Apply)
Applichiamo le modifiche ad Azure. Verrà restituito l'URL della Web App come output.

```powershell
Set-Location Day2_Lab_Terraform
terraform apply -auto-approve tfplan
```

### 8. Verifica (Idempotenza)
Rilanciando il plan, Terraform ci confermerà che l'infrastruttura corrisponde al codice e non ci sono modifiche da fare.

```powershell
Set-Location Day2_Lab_Terraform
terraform plan
```

### 9. Pulizia (Destroy) [Opzionale]
Una volta terminato il laboratorio, possiamo distruggere le risorse create da Terraform. Il Resource Group originale non verrà eliminato se gestito con un `data source`.

```powershell
# Set-Location Day2_Lab_Terraform
# terraform destroy -auto-approve
```
